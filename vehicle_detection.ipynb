```python
#building the model
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection")
trainer.setTrainConfig(object_names_array=["vehicle"], batch_size=4, num_experiments=25, 
                       train_from_pretrained_model="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/pretrained-yolov3.h5")
trainer.trainModel()
```

```python
#I rerun these code because the generated model is only 17 and my target is 20 It will not generate a model if the previous one has a lower loss. 
#I used the last generated model as the pretrained model since it has the lowest loss.
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection")
trainer.setTrainConfig(object_names_array=["vehicle"], batch_size=4, num_experiments=20, 
                       train_from_pretrained_model="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection/models/detection_model-ex-022--loss-0004.364.h5")
trainer.trainModel()
```

```python
#evaluating models
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection")
trainer.evaluateModel(model_path="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection/models", 
                      json_path="/content/drive/MyDrive/VEHICLE_DETECTION_THERMAL_IMAGES/vehicle_detection/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
```
